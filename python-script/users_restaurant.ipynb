```python
import pandas as pd
import numpy as np
```

```python
# Create an iterator that reads the file in chunks
json_iterator = pd.read_json('../data/yelp-dataset/yelp_academic_dataset_user.json', chunksize=100000, lines=True)

# Iterate over the chunks and process the data
for json_chunk_users in json_iterator:
    # Filter out rows with business IDs not in the other file
    # json_chunk_users = json_chunk_users[json_chunk_users['business_id'].isin(other_business_ids)]

    print(f"Number of rows in chunk: {len(json_chunk_users)}")
    print(json_chunk_users.head(5))
    if len(json_chunk_users) < 100000:
        break
```

```python
# Define a function to format the date string
def format_date(date_str):
    date = pd.to_datetime(date_str, format='%Y-%m-%d %H:%M:%S')
    return date.strftime('%Y%m%d')

# Apply the function to the 'date_values' column of the DataFrame and create a new column
json_chunk_users['yelping_since_values'] = json_chunk_users['yelping_since'].apply(format_date)
json_chunk_users
```

```python
# Drop the 'date' column from the DataFrame
json_chunk_users.drop(columns=['yelping_since'], inplace=True)
json_chunk_users.drop(columns=['compliment_hot','compliment_more','compliment_profile','compliment_cute','compliment_list','compliment_note','compliment_plain','compliment_cool','compliment_funny','compliment_writer','compliment_photos'], inplace=True)
json_chunk_users['review_count'] = pd.to_numeric(json_chunk_users['review_count'], errors='coerce')
json_chunk_users['review_count'] = json_chunk_users['review_count'].replace('', None)
json_chunk_users.dropna(subset=['review_count'], inplace=True)
```
