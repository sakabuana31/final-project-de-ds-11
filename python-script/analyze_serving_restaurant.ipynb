```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

# load the CSV file into a dataframe
df = pd.read_csv('../data/output-csv/serving_table_review_202303221008.csv')
# select or filter data
df = df[df['business_name'] == 'The Pancake Pantry']
df
```

```python
# Plot temperature vs. stars using scatter plot
plt.scatter(df['mean_temp'], df['stars'])
plt.xlabel('Temperature')
plt.ylabel('Stars')
plt.title('Temperature vs. Stars')
plt.show()
```

```python

# load the CSV file into a dataframe
df = pd.read_csv('../data/output-csv/serving_table_review_202303221008.csv')
# select or filter data
df = df[df['business_name'].isin(['District Donuts Sliders Brew', 'The Pancake Pantry', 'Han Dynasty', 'Prep & Pastry', 'Village Whiskey'])]

df
```

```python
# Create a pivot table with temperature and stars as the rows and columns
pivot = df.pivot_table(index='business_name', columns='stars', values='mean_temp')

# Create a heatmap of the pivot table
sns.heatmap(pivot, cmap='jet')

# Set the chart title and axis labels
plt.title('Temperature vs Stars')
plt.xlabel('Stars')
plt.ylabel('Temperature')

# Show the chart
plt.show()
```

```python
# Create a pivot table with temperature and stars as the rows and columns
pivot = df.pivot_table(index='business_name', columns='stars', values='precipitation')

# Define the new tick labels for the x and y axes
new_xticklabels = [1.0, 1.5, 2.0, 2.5]

# Create a heatmap of the pivot table
sns.heatmap(pivot, cmap='Greys', xticklabels=new_xticklabels)

# Set the chart title and axis labels
plt.title('Precipitation vs Stars')
plt.xlabel('Stars')
plt.ylabel('Precipitation')

# Show the chart
plt.show()
```
